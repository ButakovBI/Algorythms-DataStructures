```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('train_var_1.csv')
```

```python
df
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(df['Vmag'], df['e_Plx'])
plt.title('1.     : Vmag  e_Plx')
plt.xlabel('Vmag')
plt.ylabel('e_Plx')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.isnull())
plt.title('2.   ')
plt.show()
# print(len(df[df.isnull()].dropna()))
```

```python
numerical_columns = df.select_dtypes(include='number').columns
numerical_columns
```

```python
fig, axes = plt.subplots(1, len(numerical_columns), figsize=(25, 8))
for i, col in enumerate(numerical_columns):
    sns.boxplot(data=df, x=col, ax=axes[i])
    axes[i].set_title(f'Boxplot for {col}')

plt.show()
```

```python
plt.figure(figsize=(15,10))
cnt = df['TargetClass'].value_counts()
sns.barplot(x=df['TargetClass'], y=cnt)
plt.title('4.  barplot   TargetClass')
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True)

plt.title('5.      ')
plt.show()
```

```python
sns.pairplot(df[numerical_columns])
plt.suptitle('6.        ', y=1)
plt.show()
```

```python
sns.histplot(df['Vmag'])

plt.title('7.     ')
plt.show()
```
